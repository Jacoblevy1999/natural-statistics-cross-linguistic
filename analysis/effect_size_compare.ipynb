```python
%%R

effect_sizes <- read.csv("../data/SWU_effect_sizes.csv")

effect_sizes$rand_effect_size <- effect_sizes$rand_effect_size * -1
effect_sizes$max_voc_effect_size <- effect_sizes$max_voc_effect_size * -1
effect_sizes$max_turn_effect_size <- effect_sizes$max_turn_effect_size * -1

effect_sizes$rand_maxvoc_diff <- effect_sizes$max_voc_effect_size - effect_sizes$rand_effect_size
effect_sizes$rand_maxturn_diff <- effect_sizes$max_turn_effect_size - effect_sizes$rand_effect_size

effect_sizes <- effect_sizes[c("Language_name","rand_maxvoc_diff","rand_maxturn_diff")]

```

```python
%%R

library(tidyr)

effect_sizes_long_swu <- gather(effect_sizes, target_diff, effect_size_diff, rand_maxvoc_diff:rand_maxturn_diff, factor_key=TRUE)

effect_sizes_long_swu$Language_name <- as.factor(effect_sizes_long_swu$Language_name)
effect_sizes_long_swu$target_diff <- as.factor(effect_sizes_long_swu$target_diff)

effect_sizes_long_swu$nulls <- "notNaN"

effect_sizes_long_swu$nulls[is.na(effect_sizes_long_swu$effect_size_diff)] <- "NaN"

# effect_sizes_long
```

```python
%%R

library('patchwork')
library('ggpattern')
library('cowplot')
library('ggplot2')
library('scales')
library('grid')

options(repr.plot.width=6.47, repr.plot.height=7.3)

wrapper <- function(x, ...) 
{
  paste(strwrap(x, ...), collapse = "\n")
}

font_sz = 11

scalelab = "Difference in effect size"

maxvoc = "max voc"

maxturn = "max turn"

title1 = "Number of Unique Words"

title2 = "Mean Length of Utterance"

title3 = "Prop. Single Word Utterances"

p1 <- ggplot(effect_sizes_long_ld, aes(x=target_diff,y=Language_name,fill=effect_size_diff,pattern=nulls)) +
        geom_tile(data = subset(effect_sizes_long_ld, !is.na(effect_size_diff)),
                  aes(fill = effect_size_diff),
                  color = "white") +
        geom_tile_pattern(data = subset(effect_sizes_long_ld, is.na(effect_size_diff)),
                          fill = "white",
                          pattern_fill = "black",
                          pattern_angle = 45,
                          pattern_density = 0.25,
                          pattern_spacing = 0.0325,
                          pattern_key_scale_factor = 0.6) +
        scale_pattern_manual(values = c("NaN" = "stripe", "notNaN" = "none")) +
        scale_fill_gradient2(low = "firebrick1", high = "springgreen4", mid = "white",
                           midpoint = 0, limit = c(-1.2,1), oob=squish,
                           name=wrapper(scalelab, width = 15))+
        scale_y_discrete(limits = rev) +
        scale_x_discrete(position = "top",
                         labels=c(wrapper(maxvoc, width = 3),
                                  wrapper(maxturn, width = 3))) +
        coord_fixed(ratio = .3) +
        labs(tag="A",
             y = "", x = "") +
        theme_classic() +
        ggtitle(wrapper(title1, width = 15)) +
        theme(text = element_text(size=font_sz),
              axis.text.x = element_text(colour="black"),
              axis.text.y = element_text(colour="black"),
              legend.background = element_rect(fill=alpha("white",0),
                                               size=0, linetype="dotted",
                                               colour = "white"),
              legend.text=element_text(size=font_sz),
              legend.position = "left",
              legend.key.size = unit(6, 'mm'),
              legend.key.width = unit(1.5, 'mm'),
              legend.text.align = 1,
              plot.tag.position = c(.365, .975),
              plot.title = element_text(size = 11))+
        guides(pattern = FALSE)

p2 <- ggplot(effect_sizes_long_mlu, aes(x=target_diff,y=Language_name,fill=effect_size_diff,pattern=nulls)) +
        geom_tile(data = subset(effect_sizes_long_mlu, !is.na(effect_size_diff)),
                  aes(fill = effect_size_diff),
                  color = "white") +
        geom_tile_pattern(data = subset(effect_sizes_long_mlu, is.na(effect_size_diff)),
                          fill = "white",
                          pattern_fill = "black",
                          pattern_angle = 45,
                          pattern_density = 0.25,
                          pattern_spacing = 0.0325,
                          pattern_key_scale_factor = 0.6) +
        scale_pattern_manual(values = c("NaN" = "stripe", "notNaN" = "none")) +
        scale_fill_gradient2(low = "firebrick1", high = "springgreen4", mid = "white",
                           midpoint = 0, limit = c(-1.2,1), oob=squish,
                           name="Difference in effect size")+
        scale_y_discrete(limits = rev) +
        scale_x_discrete(position = "top",
                         labels=c(wrapper(maxvoc, width = 3),
                                  wrapper(maxturn, width = 3))) +
        coord_fixed(ratio = .3) +
        labs(tag="B",
             y = "", x = "") +
        theme_classic() +
        ggtitle(wrapper(title2, width = 13)) +
        theme(text = element_text(size=font_sz),
              axis.text.x = element_text(colour="black"),
              axis.text.y = element_text(colour="black"),
              legend.position = "none",
              plot.title = element_text(size = 11))

p3 <- ggplot(effect_sizes_long_swu, aes(x=target_diff,y=Language_name,fill=effect_size_diff,pattern=nulls)) +
        geom_tile(data = subset(effect_sizes_long_swu, !is.na(effect_size_diff)),
                  aes(fill = effect_size_diff),
                  color = "white") +
        geom_tile_pattern(data = subset(effect_sizes_long_swu, is.na(effect_size_diff)),
                          fill = "white",
                          pattern_fill = "black",
                          pattern_angle = 45,
                          pattern_density = 0.25,
                          pattern_spacing = 0.0325,
                          pattern_key_scale_factor = 0.6) +
        scale_fill_gradient2(low = "firebrick1", high = "springgreen4", mid = "white",
                           midpoint = 0, limit = c(-1.2,1), oob=squish,
                           name="Difference in effect size")+
        scale_y_discrete(limits = rev) +
        scale_x_discrete(position = "top",
                         labels=c(wrapper(maxvoc, width = 3),
                                  wrapper(maxturn, width = 3))) +
        coord_fixed(ratio = .3) +
        labs(tag="C",
             y = "", x = "") +
        theme_classic() +
        ggtitle(wrapper(title3, width = 17)) +
        theme(text = element_text(size=font_sz),
              axis.text.x = element_text(colour="black"),
              axis.text.y = element_text(colour="black"),
              legend.position = "none",
              plot.title = element_text(size = 11))


p <- p1 | p2 | p3

llab1 = "max has higher effect size"

llab2 = "random has higher effect size"

p <- ggdraw(p) + draw_plot_label(x = 0.1, y = .515,
                                 label = c(wrapper(llab1, width = 13)),
                                 hjust=0, size = 8, fontface = "plain")

p <- ggdraw(p) + draw_plot_label(x = 0.1, y = .395,
                                 label = c(wrapper(llab2, width = 13)),
                                 hjust=0, size = 8, fontface = "plain")

arrow_p1 <- ggplot() +
            geom_segment(aes(x = 1, y = -1, xend = 1, yend = 0),
                         col = "black",
                         arrow = arrow(length = unit(0.2, "cm"))) +
                         coord_cartesian(ylim = c(0, 0.75), xlim = c(0,50)) +
                         theme_void()
            
arrow_p2 <- ggplot() +
            geom_segment(aes(x = 1, y = 0.04, xend = 1, yend = 0),
                         col = "black",
                         arrow = arrow(length = unit(0.2, "cm"))) +
                         coord_cartesian(ylim = c(0, 0.75), xlim = c(0,50)) +
                         theme_void()

p <- ggdraw(p) +
     draw_plot(p) +
     draw_plot(arrow_p1, x = 0.025, y = 0.435) +
     draw_plot(arrow_p2, x = 0.025, y = 0.27)

ggsave("../figures/Figure_4.pdf",width = 8.5, height = 6.67, dpi=1200,p)

# p
```
